```python
df_updated=df.rename(columns={'Ship Mode':'Ship_Mode'},inplace=False)
```

```python
#Amount of money that could've been saved or converted in the profit
df_final_SC=df_updated.query('Profit<0 and Ship_Mode=="Standard Class"')
display(df_final_SC.head())
sum_SC=df_final_SC['Profit'].sum()
print("loss of profit in standard class",sum_SC)
```

```python
#Amount of money that could've been saved or converted in the profit
df_final_SEC=df_updated.query('Profit<0 and Ship_Mode=="Second Class"')
display(df_final_SEC.head())
sum_SEC=df_final_SEC['Profit'].sum()
print("loss of profit in second class",sum_SEC)
```

```python
#Amount of money that could've been saved or converted in the profit
df_final_FC=df_updated.query('Profit<0 and Ship_Mode=="First Class"')
display(df_final_FC.head())
sum_FC=df_final_FC['Profit'].sum()
print("loss of profit in First class",sum_FC)
```

```python
#Amount of money that could've been saved or converted in the profit
df_final_SD=df_updated.query('Profit<0 and Ship_Mode=="Same Day"')
display(df_final_SD.head())
sum_SD=df_final_SD['Profit'].sum()
print("loss of profit in Same day",sum_SD)
```

```python
LM=[sum_SC,sum_SEC,sum_FC,sum_SD]
x_axis=["Standard class","Second Class","First class","Same Day"]
```

```python
g=sns.scatterplot(x=x_axis,y=LM)
g.set(xlabel='Ship Mode',ylabel='Overdrawn')
```

We can see that the most overdrawn money is from the standard class even though the overall profit of every class is on the positive side but we need to focus on standard class the most becuase the money overdrawn is max for that class 

Further in the standard class Central region causes the most loss with 50th percentile of the data below 5(approx) in the mean profit while office supplies being the worst category.
